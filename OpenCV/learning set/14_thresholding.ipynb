바닥에 보석이 많이 있음. 크기에 상관없이 하나를 줍는데 1초가 걸린다고 가정.
1. 시간제한이 없다면 : 모든 보석을 전부 주울 것임.
2. 시간제한이 조금 있다면 : 크기를 내림차순하여 큰거부터 주울 것임.

시간제한은 여기서 임계값 (threshold) 역할을 함.

```python
import cv2
img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE) # 이진화 위해 흑백

ret, binary = cv2.threshold (img, 127, 255, cv2.THRESH_BINARY)

cv2.imshow ('book', img)
cv2.imshow ('ret', ret)
cv2.imshow ('binary', binary)

cv2.waitKey(0)

# mac에서 opencv 팝업창 안닫힐때 뒤에 항상 붙여주기
cv2.destroyAllWindows()
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
```

### Trackbar (값 변화에 따른 변형 확인)

```python
import cv2

def empty(pos) : 
    print (pos)
    pass

img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

name = 'Trackbar'
cv2.namedWindow (name)

cv2.createTrackbar('threshold', name, 127, 255, empty)
# bar 이름, 창의 이름, 초기값, 최댓값, 이벤트 처리 함수명

while True : # 동영상 출력과 유사
    thresh = cv2.getTrackbarPos('threshold', name) # bar 이름, 창 이름
    ret, binary = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)
    # thresh는 여기서 임계값. 우리가 트랙바를 이용하여 임계값을 조정하는 거임
    
    if not ret :
        break
    
    cv2.imshow(name, binary)
    
    if cv2.waitKey(1) == ord('q') : 
        break
    
# mac에서 opencv 팝업창 안닫힐때 뒤에 항상 붙여주기
cv2.destroyAllWindows()
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
```

## adaptive threshold

영역을 세분화하여 이진화함   
전체를 이진화할 경우, 책의 경우엔 글씨가 잘 안보일 수도 있음   
따라서 일부분만 이진화하는 것이 필요  

```python
import cv2

def empty(pos) : 
    print (pos)
    pass

img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

name = 'Trackbar'
cv2.namedWindow (name)

cv2.createTrackbar('block_size', name, 20, 100, empty) # 영역을 세분화하기 위한 블록 크기 정의 1보다는 크고 홀수만 가능)
cv2.createTrackbar('c', name, 3, 10, empty) # 양수의 값을 사용 
# bar 이름, 창의 이름, 초기값, 최댓값, 이벤트 처리 함수명

while True : # 동영상 출력과 유사
    block_size = cv2.getTrackbarPos('block_size', name) # bar 이름, 창 이름
    
    if block_size <= 1 : # 1보다는 크게끔 바꾸기
        blocksize_size = 3 # 1보다 큰 홀수
        
    if block_size % 2 == 0 : # block_size가 만약에 짝수라면, 홀수로 바꿔줘야함
        block_size += 1
        
    c = cv2.getTrackbarPos('c', name) # bar 이름, 창 이름
    
    binary = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
    # adaptive threshold는 return값이 하나
    
    cv2.imshow(name, binary)
    
    if cv2.waitKey(1) == ord('q') : 
        break
    
# mac에서 opencv 팝업창 안닫힐때 뒤에 항상 붙여주기
cv2.destroyAllWindows()
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
```

## Otsu Algo

최적의 임계값을 찾아주는 알고리즘 (트랙바 개고생 할 필요 X)

```python
import cv2
img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE) # 이진화 위해 흑백

ret, binary = cv2.threshold (img, 127, 255, cv2.THRESH_BINARY) # 일반적인 알고리즘 
ret, otsu = cv2.threshold (img, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print ('otsu : ', ret)

cv2.imshow ('book', img)
cv2.imshow ('ret', ret)
cv2.imshow ('binary', binary)
cv2.imshow ('otsu', otsu)

cv2.waitKey(0)

# mac에서 opencv 팝업창 안닫힐때 뒤에 항상 붙여주기
cv2.destroyAllWindows()
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
```
